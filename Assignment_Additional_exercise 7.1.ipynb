```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression 
%matplotlib inline
from sklearn.datasets import load_iris
iris = load_iris()


```

```python
data = pd.DataFrame(iris.data,columns=iris.feature_names)
label= pd.DataFrame(list(map(lambda x : iris.target_names[x],iris.target)), columns=['Species'])
iris= pd.concat([data, label], axis = 1)
print(iris.head())
```

###Execute the folowing Steps:
 

```python
# 1. Use the  distplot() to see the distribution of sepallengthcm, SepalwidthCm, PetalLengthCm, PetalWidthCm features. Plot them as subplots in a single image

plt.figure(1)
plt.figure(figsize = (15,10))
num_cols = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
for col in num_cols:
    plt.subplot(int(str(22)+str((num_cols.index(col)+1))))
    sns.distplot(iris[col]) 
```

```python
# 2. Do a boxplot of all feature except 'Species'
plt.figure(1)
plt.figure(figsize = (15,10))
num_cols = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
for col in num_cols:
    plt.subplot(int(str(22)+str((num_cols.index(col)+1))))
    sns.boxplot(x = col, data=iris) 
```

```python
# 3.Do a countplot for a feature 'Spesies'

iris['Species'].value_counts().plot.bar(figsize=(10,4),title='Species - Split for Dataset')
plt.xlabel('Species')
plt.ylabel('Count')
```

```python
# 4. Do a pariplot on the feature 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'Species'
sns.set()
columns = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'Species']
sns.pairplot(iris[columns],size = 3 ,kind ='scatter',diag_kind='kde')
plt.show()
```

```python
# 5. Do a lmplot on the following 'sepal length (cm)' and 'petal length (cm)'. Using hue, display the difference species in different color
sns.set()
sns.lmplot(x='sepal length (cm)', y='petal length (cm)', hue='Species', data=iris)
```

```python
# 6. Do a barplot of 'Species' vs 'sepal length (cm)'
sns.set()
sns.barplot(x='Species', y='sepal length (cm)', data=iris)
```

```python
# 7. Using heatmap, plot the correlation matrix
sns.set()
sns.heatmap(iris.corr(),cmap='Blues')
plt.title('correlation matrix')
```
